```python
def kidsWithCandies( candies, extraCandies):
        """
        :type candies: List[int]
        :type extraCandies: int
        :rtype: List[bool]
        """
        max_wc = max(candies)
        print(max_wc)
        result = []
        
        for i in candies:
            if i+extraCandies >= max_wc:
                result.append(True)
            else:
                result.append(False)
                
        return result
        
```
